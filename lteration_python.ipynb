````python
#write a program to print even number in between 500 to 600 using for loop?
for n in range (500,601):
    if (n%2 ==0):
        print(n,end=",")
````

````python
#write a program to print odd number in between 500 to 600 using for loop?
for n in range (500,601):
    if (n%2 ==0):
        print(n,end=",")
````
